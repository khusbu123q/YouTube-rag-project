```python
# Cell 1 - Imports
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from dotenv import load_dotenv
from langchain_classic.retrievers import MultiQueryRetriever
load_dotenv()
```

```python
# Cell 2 - Documents (AI & Technology topic)
all_docs = [
    Document(page_content="Deep learning models use multiple layers of neural networks to learn complex patterns from large datasets."),
    Document(page_content="Natural Language Processing enables machines to understand, interpret, and generate human language effectively."),
    Document(page_content="Reinforcement learning trains AI agents to make decisions by rewarding correct actions and penalizing mistakes."),
    Document(page_content="Transfer learning allows pre-trained models to be fine-tuned on new tasks with minimal data and computation."),
    Document(page_content="Computer vision techniques help machines analyze and understand visual data from images and videos."),
    Document(page_content="The Eiffel Tower is a famous landmark located in Paris, France, built in 1889."),
    Document(page_content="Football is one of the most popular sports in the world with billions of fans globally."),
    Document(page_content="Photosynthesis is the process by which plants use sunlight to produce food from carbon dioxide and water."),
    Document(page_content="The stock market allows investors to buy and sell shares of publicly listed companies."),
]
```

```python
# Cell 3 - Embedding Model and Vector Store
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create FAISS vector store
vector_store = FAISS.from_documents(documents=all_docs, embedding=embedding)
```

```python
# Cell 4 - Simple Retriever
similarity_retriever = vector_store.as_retriever(search_type='similarity', search_kwargs={'k': 5})
```

```python
# Cell 5 - LLM and MultiQuery Retriever
llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation"
)
model = ChatHuggingFace(llm=llm)

# MultiQuery Retriever
multiquery_retriever = MultiQueryRetriever.from_llm(
    retriever=vector_store.as_retriever(search_kwargs={'k': 5}),
    llm=model
)
```

```python
# Cell 5 - LLM and MultiQuery Retriever
llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation"
)
model = ChatHuggingFace(llm=llm)

# MultiQuery Retriever
multiquery_retriever = MultiQueryRetriever.from_llm(
    retriever=vector_store.as_retriever(search_kwargs={'k': 5}),
    llm=model
)
```

```python
# Cell 6 - Query
query = 'How do machines learn and understand data patterns?'
```

```python
# Cell 7 - Get Results
similarity_results = similarity_retriever.invoke(query)
multiquery_results = multiquery_retriever.invoke(query)
top_n = 5
multiquery_results = multiquery_results[:top_n]
```

```python
# Cell 8 - Print Similarity Results
for i, doc in enumerate(similarity_results):
    print(f"\n ----Result {i+1}---- ")
    print(f"Content:\n{doc.page_content}---")
```

```python
# Cell 9 - Print MultiQuery Results
for i, doc in enumerate(multiquery_results):
    print(f"\n ----Result {i+1}---- ")
    print(f"Content:\n{doc.page_content}---")
```
